```python
import yfinance as yf
```

```python
# Création de l'objet pour une action (ex : Apple)
apple = yf.Ticker("AAPL,AMZN,GOOGL")  # Plusieurs actions peuvent être spécifiées

# Récupération des données historiques
data = apple.history(start="2024-01-01", end="2025-01-01")  # 1 an d'historique
data
```

```python
data.rename(columns={
            'Close': 'Price'
        }).reset_index().to_dict(orient='records')
```
